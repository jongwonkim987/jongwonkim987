[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def sum_numbers(*args):
    total = 0
    for nums in args:
        total += nums
    return total

print(sum_numbers(20,40534,123840,193484))
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n+1):
        if i % 2 !=0:
            yield i

for num in odd_numbers(10):
    print(num)
print(list(odd_numbers(10)))
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(nums):
    if nums % 2 == 0:
        return True
    return False

print(is_even(6))
print(is_even(5))
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

factorial(5)
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(string):
    for i in string:
        print(len(i))

str_list = ["words", "banana", "python", "task"]
print_lengths(str_list)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(string1, string2):
    if len(string1) >= len(string2):
        return string1
    else:
        return string2

print(longer_string("apple", "hi"))
print(longer_string("dog", "cat"))
print(longer_string("hi", "hello"))
print(longer_string("abc", "def"))
```
