[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
nums_list = []

while True:
    user_input = input("숫자를 입력해주세요. (done입력시 종료)")
    if user_input == "done":
        break
    try:
        num = int(user_input)
        nums_list.append(num)
    except ValueError:
        print(f'잘못된 값: {user_input}, 숫자를 입력해주세요.')

print("입력된 숫자 목록:", nums_list)
````

requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>

(실습 전 requests 모듈을 설치해야 함)

````python
import requests #requests -> 웹사이트에서 http 요청을 보내서 데이터를 받아오는 라이브러리

url = "https://kream.co.kr/"
try:
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; simple-scraper/1.0)"
    }
    response = requests.get(url, headers=headers, timeout=10)
    print("Status Code:", response.status_code)
    print("HTML (first 100 chars):")
    print(response.text[:100])
except requests.exceptions.RequestException as e: #요청관련 오류를 한번에 처리
    print("요청 중 오류가 발생했습니다", e)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time
from functools import wraps

def timer_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f'함수 실행 시간: {end - start:.2f}초')
        return result
    return wrapper

@timer_decorator
def test():
    time.sleep(1.2)
    return "너무 어려움"
print(test())
````

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
import math

def triangle_area(base, height):
    return (base * height) / 2
def circle_area(radius):
    return math.pi * radius * radius
def box_volume(width, height, depth):
    return width * height * depth
````

````python
import mymath

tri = mymath.triangle_area(10, 5)
print("삼각형의 넓이:", tri)

cir = mymath.circle_area(3)
print("원의 넓이:", cir)

box = mymath.box_volume(2, 3, 4)
print("직육면체의 넓이:", box)
````

[문제 9]

animals라는 패키지를 만들고, 이 패키지 내에 mammals, birds라는 두 개의 서브 모듈을 생성하세요. 

- 각 모듈에는 최소한 한 개 이상의 동물에 관한 클래스를 정의하세요<br>
  (예: mammals에는 Dog, birds에는 Eagle)
- 이 패키지와 모듈을 사용하여 동물들의 정보를 출력하는 프로그램을 작성

폴더 및 파일 구조
```
animals/(폴더)
├── __init__.py
├── mammals.py
└── birds.py
```


````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def info(self):
        return f'강아지 이름: {self.name}, 나이: {self.age}살'
````

````python
class Eagle:
    def __init__(self, species):
        self.species = species
    def info(self):
        return f'독수리 종류: {self.species}'
````

````python
from animals.manmals import Dog
from animals.birds import Eagle

dog = Dog("바둑이", 3)
eagle = Eagle("대머리독수리")

print(dog.info())
print(eagle.info())
````
