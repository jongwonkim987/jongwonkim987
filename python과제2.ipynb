__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
merge = {"strengh":10, "agility":20, "intelligence":80, "stamina":50}
warrior = {"strengh":80, "agility":30, "intelligence":20, "stamina":100}  #앞으로의 편의를 위해 미리 작성
rogue = {"strengh":40, "agility":80, "intelligence":30, "stamina":60}     #내장 함수와 겹칠수도 있어서 혹시 몰라 "str","int"로 줄이지 않음
healer = {"strengh":20, "agility":40, "intelligence":70, "stamina":70}    #과제를 다하고 ""안에 내용을 문자로 인식하기 때문에 사용해도 상관없었겠다는 생각이 듬
if merge["intelligence"] >= 50:
    print(f'현재 지능: {merge["intelligence"]}이므로 50이상입니다.')
else:
    print("지능이 50 미만입니다.")
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
rogue = {"strengh":40, "agility":80, "intelligence":30, "stamina":60}
if rogue["agility"] >=70:
    print(f'현재 민첩성: {rogue["agility"]}이므로 70이상입니다.')
else:
    print("민첩성이 70 미만입니다.")
if rogue["strengh"] >=50:
    print(f'현재 힘: {rogue["strengh"]}이므로 50이상입니다.')
else:
    print("힘이 50 미만입니다.")
```

__[문제 4: 네 번째 사냥]__

```python
healer = {"strengh":20, "agility":40, "intelligence":70, "stamina":70}
if healer["intelligence"] >=60:
    print(f'현재 지능: {healer["intelligence"]}이므로 60이상입니다.')
else:
    print("지능이 60 미만입니다.")
if healer["stamina"] >=50:
    print(f'현재 체력: {healer["stamina"]}이므로 50이상입니다.')
else:
    print("체력이 50 미만입니다.")
```

__[문제 5: 레벨업]__

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
merge_ex, warrior_ex, rogue_ex, healer_ex = 120, 150, 110, 90
if merge_ex >=100 :
    print(f'마법사의 현재 경험치: {merge_ex}이므로 100이상입니다.')  #너무 지저분해서 구글링을 해보았는데
else:                                                                #for문을 이용하라고 함 -> 아직 학습하지 않음 -> 오늘 vod보고 for문으로 다시 도전
    print("마법사의 경험치가 100 미만입니다.")
if warrior_ex >=100 :
    print(f'전사의 현재 경험치: {warrior_ex}이므로 100이상입니다.')
else:
    print("전사의 경험치가 100 미만입니다.")
if rogue_ex >=100 :
    print(f'도적의 현재 경험치: {rogue_ex}이므로 100이상입니다.')
else:
    print("도적의 경험치가 100 미만입니다.")
if healer_ex >=100 :
    print(f'힐러의 현재 경험치: {healer_ex}이므로 100이상입니다.')
else:
    print("힐러의 경험치가 100 미만입니다.")
```
