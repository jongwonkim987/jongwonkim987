__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
merge_str, merge_agi, merge_int, merge_sta = 10, 20, 80, 50
print(f'지능이 50이상인가?: {merge_int>=50}')
```

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다. 

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
warrior_str, warrior_agi, warrior_int, warrior_sta = 80, 30, 20, 100
warrior_att = warrior_str >= 60 and warrior_sta >= 80
print(f'전사의 힘이 60이상이고 체력이 80이상인가?: {warrior_att}')
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
rogue_str, rogue_agi, rogue_int, rogue_sta = 40, 80, 30, 60
roque_att = rogue_agi >= 70 or rogue_str >= 50
print(f'도적의 민첩성이 70이상 이거나 힘이 50이상인가?: {roque_att}')
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
merge_xp, warrior_xp, rogue_xp, healer_xp = 120, 150, 110, 90
level_up = { "merge": merge_xp >= 100, "warrior": warrior_xp >= 100, "rogue": rogue_xp >= 100, "healer": healer_xp >= 100}
print(f'마법사 경험치가 100이상이면 레벨업: {level_up["merge"]}')
print(f'전사 경험치가 100이상이면 레벨업: {level_up["warrior"]}')
print(f'도적 경험치가 100이상이면 레벨업: {level_up["rogue"]}')
print(f'치유사 경험치가 100이상이면 레벨업: {level_up["healer"]}')
```

```python
party_xp = {"merge": 120, "warrior": 150, "rogue": 110, "healer": 90}
for role, xp in party_xp.items():
    level_up = xp >= 100
    print(f'{role} 경험치가 100이상이면 레벨업: {level_up}')
```
